```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# 1. Oxford University Faculty Data

```python
# Fetch the page
response = requests.get('https://www.cs.ox.ac.uk/people/faculty.html')
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <li> elements
li = soup.find_all('li')

# Store data
faculty = []
university = 'University of Oxford'

# Extract names and roles
for i in li:
    spans = i.find_all('span')
    if len(spans) == 2:
        name = spans[0].text.replace('\xa0', ' ')
        role = spans[1].text
        faculty.append({'Name': name, 'Role': role, 'University': university})

# Create DataFrame
df1 = pd.DataFrame(faculty)

df1.to_csv('Oxford_University_Faculty_Data.csv', index=False)
df1.sample(5)
```

# 2. Cambridge University Faculty Data

```python
response = requests.get('https://www.cst.cam.ac.uk/people/directory')
soup = BeautifulSoup(response.text, 'html.parser')
faculty = []
university = 'University of Cambridge'
table_body = soup.find_all('tbody')
for i in table_body:
  t_d = i.find_all('td')
  for i in range(0,len(t_d),5):
    name = t_d[i].text.strip()
    id = t_d[i+1].text.strip()
    office_phone = t_d[i+2].text.strip()
    room = t_d[i+3].text.strip()
    role = t_d[i+4].text.strip()
    faculty.append({'Name':name,'Role':role,'University': university,'ID':id,'Office Phone':office_phone,'Room':room})
df2 = pd.DataFrame(faculty)
df2.to_csv('Cambridge_University_Faculty_Data.csv', index=False)
df2.sample(5)
```

# 3. Imperial College London Faculty Data


```python
# Step 1: Fetch the web page
url = 'https://www.imperial.ac.uk/computing/people/academic-staff/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 2: Prepare lists for data
faculty = []
university = 'Imperial College London'

# Step 3: Locate people container
people_ul = soup.find_all('ul', class_='people')

# Step 4: Loop through each person and extract info
for ul in people_ul:
    li_tags = ul.find_all('li')
    for n in li_tags:
        # Name
        name = n.find('span', class_='person-name')
        name = name.text.strip() if name else ""

        # Job Title
        job = n.find('span', class_='job-title')
        role = job.text.strip() if job else ""

        # Phone
        phone = n.find('span', class_='tel')
        phone = phone.text.strip() if phone else ""

        # Address
        address_block = n.find('div', class_='address-wrapper')
        address_text = ""
        if address_block:
            lines = address_block.find_all('p')
            address_text = " ".join(line.text.strip() for line in lines)

        # Research Interests
        interest_block = n.find('div', class_='dept-wrapper')
        interest_text = ""
        if interest_block:
            lines = interest_block.find_all('p')
            interest_text = " ".join(line.text.strip() for line in lines)

        faculty.append({'Name':name,'Role':role,'University': university,'Phone': phone,'Address': address_text,'Research Interests': interest_text})


d3 = pd.DataFrame(faculty)
d3.to_csv('Imperial_College_London_Faculty_Data.csv', index=False)
d3.sample(5)
```

# 4. University of Edinburgh (Academic Staff)

```python
response = requests.get('https://people.inf.ed.ac.uk/academic.html')
soup = BeautifulSoup(response.text, 'html.parser')
faculty = []
university = 'University of Edinburgh'
tdata = soup.find_all('td')
for i in range(0,len(tdata),5):
  name = tdata[i].text.strip()
  role = tdata[i+1].text.strip()
  email = tdata[i+2].text.strip()
  room = tdata[i+3].text.strip()
  phone = tdata[i+4].text.strip()
  faculty.append({'Name':name,'Role':role,'University': university,'Email': email,'Room': room,'Phone': phone})
df4 = pd.DataFrame(faculty)
df4.to_csv('University_of_Edinburgh_Faculty_Data.csv', index=False)
df4.sample(5)
```

# 5. University of Manchester

```python
webpage=requests.get('https://www.cs.manchester.ac.uk/about/people/academic-and-research-staff/').text
soup=BeautifulSoup(webpage,'lxml')
faculty = []
university = 'University of Manchester'

ul = soup.find_all('ul',class_='tabRow')
for i in ul:
  li = i.find_all('li')
  for i in li:
    div = i.find_all('div',class_='tabCol_30')
    name = div[0].text.replace("Name: ","").strip()
    role = div[1].text.replace("Role: ","").strip()
    if len(div) >= 3:
      area_of_expertise = div[2].text.replace("Area of Expertise: ", "").strip()
    else:
      "N/A"
    faculty.append({'Name':name,'Role':role,'University': university,'Area of Expertise': area_of_expertise})
df5 = pd.DataFrame(faculty)
df5.to_csv('University_of_Manchester_Faculty_Data.csv', index=False)
df5.sample(5)
```

# 6. University of Warwick

```python
webpage=requests.get('https://warwick.ac.uk/fac/arts/history/people/staff_index/').text
soup=BeautifulSoup(webpage,'lxml')
faculty = []
university = 'University of Warwick'
tdata = soup.find_all('td')
for i in tdata:
  h4 = i.find_all('h4')
  for i in h4:
    h4_text = i.text.strip()
    name, role = h4_text.split('-', 1)
    faculty.append({'Name':name.strip(),'Role':role.strip(),'University': university})
df6 = pd.DataFrame(faculty)
df6.to_csv('University_of_Warwick_Faculty_Data.csv', index=False)
df6.sample(5)
```

# 7. University of Surrey

```python
webpage = requests.get('https://www.surrey.ac.uk/school-computer-science-and-electronic-engineering/people/academic-staff').text
soup = BeautifulSoup(webpage, 'lxml')

faculty = []
university = 'University of Surrey'

profiles = soup.find_all('div', class_='paragraph--type--university-profile')

for profile in profiles:
    name_tag = profile.find('h3', class_='font-proxima-nova-semibold')
    name = name_tag.text.strip()

    p_tags = profile.find_all('p')
    for p in p_tags:
      role = p.text.strip()

    faculty.append({
        'Name': name,
        'Role': role,
        'University': university
    })

df7 = pd.DataFrame(faculty)
df7.to_csv('University_of_Surrey_Faculty_Data.csv', index=False)
df7.sample(5)
```

# 8. University of Leicester


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

faculty = []
university = 'University of Leicester'
total_pages = 160

for page in range(total_pages):
    url = f'https://le.ac.uk/people?Q=&Page={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    cards = soup.find_all('li', class_='people-results__item')

    for card in cards:
        name_tag = card.find('a', class_='people-results__link--title')
        role_tag = card.find('p', class_='people-results__position')

        if name_tag:  # Only proceed if name exists
            name = ' '.join(name_tag.get_text().split()) if name_tag else 'N/A'
            role = role_tag.get_text(strip=True) if role_tag else 'N/A'
            faculty.append({
                'Name': name,
                'Role': role,
                'University': university
            })

# Convert to DataFrame
df8 = pd.DataFrame(faculty)
df8.to_csv('University_of_Leicester_Faculty_Data.csv', index=False)
df8.sample(5)
```

# 9. Loughborough University London

```python
webpage = requests.get('https://www.lborolondon.ac.uk/staff/').text
soup = BeautifulSoup(webpage, 'lxml')

faculty = []
university = 'Loughborough University London'

div = soup.find_all('div', class_='list__name-role')
for i in div:
  name_tags = i.find_all('span',class_='list__link-text')
  role_tags = i.find_all('span',class_='list__role')

  for i in range(len(name_tags)):
    name = name_tags[i].text.strip()
    role = role_tags[i].text.strip()
    faculty.append({'Name':name,'Role':role,'University': university})
df9 = pd.DataFrame(faculty)
df9.to_csv('Loughborough_University_London_Faculty_Data.csv', index=False)
df9.sample(5)
```
